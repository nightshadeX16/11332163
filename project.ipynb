```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('red_wine.csv')
dataset
```

```python
class Data_Extractor:
    def __init__(self, dataframe):
        self.dataframe = dataframe

    def view_first_n_rows(self, n):
        return self.dataframe.head(n)

    def view_last_n_rows(self, x):
        return self.dataframe.tail(x)

    def dataset_info(self):
        return self.dataframe.info()

    def dataset_description(self):
        return self.dataframe.describe()

    def null_value_checker(self, column_name):
        print(f"There are {self.dataframe[column_name].isnull().sum()} null values in the '{column_name}' column")
```

```python
class Data_Modifier:
    def __init__(self, dataframe):
        self.dataframe = dataframe

    def replace_nulls_with_mean(self, column_name):
            if column_name == 'fixed acidity':
                n = 1
            elif column_name == 'volatile acidity':
                n = 3
            elif column_name == 'citric acid':
                n = 2
            elif column_name == 'chlorides':
                n = 3
            elif column_name == 'free sulfur dioxide':
                n = 1
            elif column_name == 'density':
                n = 5
            else:
                n = 1
            self.dataframe[column_name].fillna(self.dataframe[column_name].mean().round(n), inplace=True)

    def replace_nulls_with_median(self, column_name):
            if column_name == 'fixed acidity':
                n = 1
            elif column_name == 'volatile acidity':
                n = 3
            elif column_name == 'citric acid':
                n = 2
            elif column_name == 'chlorides':
                n = 3
            elif column_name == 'free sulfur dioxide':
                n = 1
            elif column_name == 'density':
                n = 5
            else:
                n = 1
            self.dataframe[column_name].fillna(self.dataframe[column_name].median().round(n), inplace=True)
```

```python
class Red_Wine_Dataset(Data_Extractor, Data_Modifier):
    def __init__(self, dataframe):
        super().__init__(dataframe)
```

```python
data = Red_Wine_Dataset(dataframe=dataset)
```

```python
# View the first 10 rows of the dataset
data.view_first_n_rows(10)
```

```python
# View the last 20 rows of the dataset
data.view_last_n_rows(20)
```

```python
# Checking basic info on the dataset
data.dataset_info()
```

```python
# Obtaining statistical information on the dataset
data.dataset_description()
```

```python
# Data Preprocessing

data.replace_nulls_with_mean('fixed acidity')
data.replace_nulls_with_mean('volatile acidity')
data.replace_nulls_with_mean('citric acid')
data.replace_nulls_with_mean('chlorides')
data.replace_nulls_with_median('free sulfur dioxide')
data.replace_nulls_with_mean('density')
```

```python
# Verifying data preprocess
data.dataset_info()
```

```python
# Histogram for fixed acidity

plt.figure(figsize=(6, 6))
plt.hist(dataset['fixed acidity'], bins=20, color='skyblue', edgecolor='black', density=True)
plt.title('Histogram of Fixed Acidity (Figure 1)')
plt.xlabel('Fixed Acidity')
plt.ylabel('Frequency')
plt.show()
```

```python
data = pd.DataFrame({
    'fixed acidity': dataset['fixed acidity'],
    'volatile acidity': dataset['volatile acidity'],
    'citric acid': dataset['citric acid'],
    'residual sugar': dataset['residual sugar'],
    'chlorides': dataset['chlorides']
})

# Create a pair plot
fig, axes = plt.subplots(nrows=5, ncols=5, figsize=(12, 12))

# Histogram plots on the diagonal
for i, col in enumerate(data.columns):
    axes[i, i].hist(data[col], bins=20, color='skyblue', edgecolor='black')
    axes[i, i].set_title(col)

# Lower triangle : Scatter plots
for i in range(1, 5):
    for j in range(i):
        axes[i, j].scatter(data[data.columns[j]], data[data.columns[i]], alpha=0.5, color='coral')
        axes[i, j].set_xlabel(data.columns[j])
        axes[i, j].set_ylabel(data.columns[i])

# Upper triangle : 2D Histograms (or kernel density estimates/heatmap)
for i in range(1, 5):
    for j in range(i):
        axes[j, i].hist2d(data[data.columns[j]], data[data.columns[i]], bins=20, cmap='Reds')
        axes[j, i].set_xlabel(data.columns[j])
        axes[j, i].set_ylabel(data.columns[i])

plt.tight_layout()
plt.show()
```

```python
data = pd.DataFrame({
    'quality': dataset['quality']
})

# Count occurrences of each quality level
quality_counts = data['quality'].value_counts()

# Create a pie chart
plt.figure(figsize=(8, 8))
colors = plt.cm.Paired(range(len(quality_counts)))  # Use a colormap for different colors
plt.pie(quality_counts, labels=quality_counts.index, autopct='%1.1f%%', colors=colors, startangle=140)
plt.title('Distribution of Wine Quality (Figure 3)')
plt.show()
```

```python
data = pd.DataFrame({
    'fixed acidity': dataset['fixed acidity'],
    'volatile acidity': dataset['volatile acidity'],
    'citric acid': dataset['citric acid'],
    'residual sugar': dataset['residual sugar'],
    'chlorides': dataset['chlorides']
})

# Calculate correlation matrix
correlation_matrix = data.corr()

# Create a correlation heatmap using matplotlib
plt.figure(figsize=(10, 8))
heatmap = plt.imshow(correlation_matrix, cmap='coolwarm', interpolation='none')
plt.colorbar(heatmap, fraction=0.046, pad=0.04)  # Add color bar

# Display correlation values in the heatmap
for i in range(len(correlation_matrix)):
    for j in range(len(correlation_matrix)):
        text = plt.text(j, i, f"{correlation_matrix.iloc[i, j]:.2f}",
                       ha='center', va='center', color='black', fontsize=10)

# Set axis labels
plt.xticks(np.arange(len(correlation_matrix)), correlation_matrix.columns, rotation=45, ha='right')
plt.yticks(np.arange(len(correlation_matrix)), correlation_matrix.index)

# Set plot title
plt.title('Correlation Heatmap (Figure 5)')

# Show the plot
plt.show()
```
